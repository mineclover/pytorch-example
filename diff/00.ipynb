```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='TRUE'
os.environ['CUDA_LAUNCH_BLOCKING']='1'
import torch
import numpy as np
import matplotlib.pyplot as plt
# GPU 사용 가능하면 cuda, 아니면 cpu
import torchvision
import torchvision.transforms as transforms
from torch import Tensor
from typing import List, Tuple, Dict, Any, Union


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.cuda.empty_cache()
```

```python
from torch import nn
# 간단하게 그래프 비교해보고 가기

# x축을 -10 ~ 10 까지를 100눈금으로 나누어서 설정
x = torch.linspace(-3, 3, 100)

# ReLU 함수
relu_fn = nn.ReLU()
relu = relu_fn(x)

# PyTorch Tensor -> Numpy ndarray
plt.plot(x.numpy(), relu.numpy(), label='ReLU')


# 원점 분기로 나뉘는 것 시각화를 위해 추가
plt.axhline(0, color='red', linewidth=0.3)
plt.axvline(0, color='red', linewidth=0.3)

plt.title("ReLU vs LeakyReLU")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from torch import nn
# 간단하게 그래프 비교해보고 가기

# x축을 -10 ~ 10 까지를 100눈금으로 나누어서 설정
x = torch.linspace(-3, 3, 100)

# LeakyReLU 함수
leaky_fn = nn.LeakyReLU(negative_slope=0.2)
leaky = leaky_fn(x)

# PyTorch Tensor -> Numpy ndarray
plt.plot(x.numpy(), leaky.numpy(), label='LeakyReLU (α=0.01)')

# 원점 분기로 나뉘는 것 시각화를 위해 추가
plt.axhline(0, color='red', linewidth=0.3)
plt.axvline(0, color='red', linewidth=0.3)

plt.legend()
plt.grid(True)
plt.show()
```
